```python
import cv2
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
current_dir = os.getcwd()
image_dir = current_dir + "/images/" + "sudoku_2.png"
```

```python
image = cv2.imread(image_dir)
plt.imshow(image)
```

```python
# image = cv2.resize(image, (900, 900))
# plt.imshow(image)
```

```python
def preprocess(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (3,3),6) 
    thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 199, 5)
    plt.imshow(thresh)
    return thresh

thresh_image = preprocess(image)
print(thresh_image)
```

```python
contour_1 = thresh_image.copy()
contour_2 = thresh_image.copy()
contour_3 = thresh_image.copy()
edged = cv2.Canny(contour_1, 30, 200) 
contours, heirarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print(len(contours))
for contour in contours:
    print(len(contour))
cv2.drawContours(contour_2, contours, -1, (0, 255, 0), 3)
cv2.drawContours(contour_3, contours, 4, (0, 255, 255), 3)
plt.imshow(contour_2)
plt.imshow(contour_3)
print(contours[4])
```

```python
largest_contour = max(contours, key=cv2.contourArea)

epsilon = 0.02 * cv2.arcLength(largest_contour, True)
approx = cv2.approxPolyDP(largest_contour, epsilon, True)
print("approx", approx)

if len(approx) == 4:
    box = approx.reshape(4, 2)
    print("box", box)
else:
    raise ValueError("The largest contour does not have four corners.")


# top-left, top-right, bottom-right, bottom-left
def order_points(pts):
    rect = np.zeros((4, 2), dtype="float32")
    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]
    return rect

ordered_box = order_points(box)

dst = np.float32([[0, 0], [900, 0], [900, 900], [0, 900]])
matrix = cv2.getPerspectiveTransform(ordered_box, dst)
result = cv2.warpPerspective(image, matrix, (900, 900))

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
```

```python
preprocess(result)
```
